```python
import re
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer, ENGLISH_STOP_WORDS
from sklearn.metrics.pairwise import cosine_similarity
from nltk.stem import PorterStemmer
```

1. Create Dataset

```python
movies = [
    {
        "title": "Galaxy Wars",
        "description": "A space adventure where humans and robots fight to save the galaxy."
    },
    {
        "title": "Robot Planet",
        "description": "A young scientist discovers a planet full of intelligent robots."
    },
    {
        "title": "Healthy Life",
        "description": "A documentary about healthy food, exercise, and balanced lifestyle."
    },
    {
        "title": "Food Journey",
        "description": "A chef travels around the world to explore healthy and traditional food."
    },
    {
        "title": "Love in Paris",
        "description": "A romantic story about two people who meet in Paris."
    },
    {
        "title": "Football Dreams",
        "description": "A young player works hard to become a professional football star."
    },
    {
        "title": "Ocean Mystery",
        "description": "A group of divers search for hidden treasure under the ocean."
    },
    {
        "title": "Alien Attack",
        "description": "Aliens invade Earth and a brave team must protect humanity."
    },
    {
        "title": "Cyber Future",
        "description": "In the future, robots and artificial intelligence control the city."
    },
    {
        "title": "Jungle Escape",
        "description": "Friends get lost in a jungle and try to survive dangerous animals."
    }
]

df = pd.DataFrame(movies)
```

2. Preprocessing Function

```python
stemmer = PorterStemmer()

def preprocess_text(text):
    text = text.lower()                         # lowercase
    text = re.sub(r"[^a-z\s]", "", text)        # remove punctuation/numbers
    words = text.split()                        # tokenization

    clean_words = []
    for word in words:
        if word not in ENGLISH_STOP_WORDS:      # remove stopwords
            stemmed_word = stemmer.stem(word)   # stemming
            clean_words.append(stemmed_word)

    return " ".join(clean_words)
```

```python
# Combining title + description
df["combined_text"] = df["title"] + " " + df["description"]

# Applying preprocessing
df["clean_text"] = df["combined_text"].apply(preprocess_text)
```

3. TF-IDF Vectorization

```python
vectorizer = TfidfVectorizer()

tfidf_matrix = vectorizer.fit_transform(df["clean_text"])
```

 4. Accept User Query

```python
user_query = input("Enter your search query: ")

clean_query = preprocess_text(user_query)

query_vector = vectorizer.transform([clean_query])
```

5. Cosine Similarity

```python
similarity_scores = cosine_similarity(query_vector, tfidf_matrix).flatten()

df["similarity_score"] = similarity_scores
```

6. Return Top 3 Results

```python
top_3 = df.sort_values(by="similarity_score", ascending=False).head(3)

print("\nUser Query:", user_query)
print("\nTop 3 Recommended Movies:\n")

for index, row in top_3.iterrows():
    print("Title:", row["title"])
    print("Description:", row["description"])
    print("Similarity Score:", round(row["similarity_score"], 3))
    print("-" * 50)
```
